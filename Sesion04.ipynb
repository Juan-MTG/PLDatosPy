```python
df2 = pd.read_json("datos/alumnos_calif.json")
df2
```

```python
#Agrupar por sexo, no cuenta los nulos
df2.groupby(['sexo'])['clave_alu'].agg('count')
```

```python
#Agrupar por sexo, contando los nulos
df2.groupby(['sexo'],dropna = False)['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad, no cuenta los nulos
df2.groupby(['ciudad'])['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad, contando los nulos
df2.groupby(['ciudad'],dropna = False)['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad y sexo, cuenta los nulos
df2.groupby(['ciudad', 'sexo'], dropna=False)['clave_alu'].agg('count')
```

```python
#Agrupar por ciudad y sexo, cuenta los nulos
df2.groupby(['ciudad', 'sexo'], dropna=False)['calificacion']\
    .agg(['count','sum', 'mean','min', 'max', 'std'])
```

```python
#Agrupar por ciudad y sexo, cuenta los nulos
#RESETEAMOS LOS INDICES PARA PODER SER GUARDADO
df_e = df2.groupby(['ciudad', 'sexo'], dropna=False)['calificacion']\
    .agg(['count','sum', 'mean','min', 'max', 'std']).reset_index()
df_e
```

```python
#Agrupar por sexo, obtener promedio
df2.groupby('sexo', dropna=False)['calificacion'].mean()
```

```python
#Agrupar por sexo, obtener promedio, filtrado por la cd de QUERETARO
df2[df2['ciudad']=='QUERETARO']\
    .groupby('sexo', dropna=False)['calificacion']\
    .agg(['count','mean'])
```

```python
#Agrupar por sexo, obtener promedio, filtrado por la cd de QUERETARO
df2.groupby('sexo', dropna=False)['calificacion']\
    .agg(['count','mean'])
```

```python
#Crear una nueva columna
df2['calificacion_cadena'] = df2['calificacion'].astype(str)
df2.info()
```

```python
df2[['entero','decimal']] = df2.calificacion_cadena.str.split('.', expand=True)
df2.drop('calificacion_cadena', axis=1, inplace=True)
df2
```

```python
df2.calificacion_cadena.str.split('.', expand=True)
```

```python
# El GPA (Grade Point Average) es un término utilizado para asignar un valor numérico 
# a las calificaciones acumuladas 
df2['gpa'] = (((df2['calificacion']*100)/df2['calificacion'].max())/20) -1
df2
```

```python
df2[['usuario','dominio']] = df2.email.str.split('@', expand=True)
df2
```

```python
#Generar valor escalar
gpa_rango = [1, 1.7, 2.4, 3.3, 4.1]
gpa_etiqueta = ['D', 'C', 'B', 'A']
df2['gpa_escala'] = pd.cut(df2['gpa'], bins=gpa_rango, labels=gpa_etiqueta)
print(df2['gpa_escala'].unique())
df2.info()
```
