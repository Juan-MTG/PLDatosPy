```python
import pandas as pd
```

```python
df = pd.read_json("datos/alumnos_calif.json")
df
```

```python
df['estado'] = df.curp.str[11:13]
df['gpa'] = (((df['calificacion']*100)/df['calificacion'].max())/20) -1
gpa_rango = [1, 1.7, 2.4, 3.3, 4.1]
gpa_etiqueta = ['D', 'C', 'B', 'A']
df['gpa_escala'] = pd.cut(df['gpa'], bins=gpa_rango, labels=gpa_etiqueta)
sti = pd.date_range('2023-01-01', periods=df.clave_alu.count(), freq='2m')
df['fecha'] = sti
df
```

```python
#Cambiar NaN por un valor
df_na['ciudad'] = df_na.ciudad.fillna('SIN CD')
df_na
```

```python
#Cambiar NaN por el promedio de calificaci√≥n
df_nac = df_na.copy()
df_nac['calificacion'] = round(df_na.calificacion.fillna(df_na.calificacion.mean()),1)
df_nac
```

```python
#Cambiar NaN por el m√≠nimo de calificaci√≥n
df_nac = df_na.copy()
df_nac['calificacion'] = round(df_na.calificacion.fillna(df_na.calificacion.min()),1)
df_nac
```

```python
#Remplazar valores
df['ciudad'] = df['ciudad'].replace({'SANTIAGO DE QUERETARO':'QUERETARO'})
df['ciudad'] = df['ciudad'].replace({'QUERETARO QRO.':'QUERETARO'})
df
```

```python
df_estado = pd.read_csv("datos/estados_curp.csv")
df_estado
```

```python
#Inner cuando las 2 claves coinciden
pd.merge(df,df_estado,left_on='estado',right_on='ABREVIATURA',how='inner')
```

```python
#LEFT muestra todo lo de la izquierda (ALUMNOS) aunque no coincida la llave
pd.merge(df,df_estado,left_on='estado',right_on='ABREVIATURA',how='left')
```

```python
#RIGHT muestra todo lo de la derecha (ESTADOS) aunque no coincida la llave
pd.merge(df,df_estado,left_on='estado',right_on='ABREVIATURA',how='right')
```

```python
#OUTER muestra todo lo de la izquierda (ALUMNOS) y todo lo de la derecha (ESTADOS) aunque no coincida la llave
pd.merge(df,df_estado,left_on='estado',right_on='ABREVIATURA',how='outer')
```
